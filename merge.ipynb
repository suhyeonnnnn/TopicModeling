# RSN 다시 MERGE

```python
import os
file_list = os.listdir('C:\\Users\\sh921\\Documents\\카카오톡 받은 파일\\RSNGAM~2\\RSN game_20190117-20210120\\')
file_list
```

```python
os.chdir("C:\\Users\\sh921\\Documents\\카카오톡 받은 파일\\RSNGAM~2\\RSN game_20190117-20210120")
print(os.getcwd())
# /Users
```

```python
import pandas as pd

for file in file_list:
    # 첫번째 시트를 읽어옵니다.
    df1 = pd.read_excel(file, sheet_name = '긍정')
    # 두번째 시트를 읽어옵니다.
    df2 = pd.read_excel(file, sheet_name = '부정')
    df3 = pd.read_excel(file, sheet_name = '중립')
    concatted_df = pd.concat([df1,df2,df3], axis=1)
    concatted_df['gameCode'] = file 
    concatted_df.to_csv('{}.csv'.format(file))
```
