```python
date = datetime. strptime("2020-01-01", '%Y-%m-%d')
new = pd.DataFrame()
timelst = []
avgtime = []
user = []
datelst = []
for i in range(365):
    datedf = nf[nf.date == date]
    datelst.append(date)
    timelst.append(datedf['timeCountTotal'].sum())
    avgtime.append(datedf['avgUsedTime'].sum())
    user.append(datedf['userCountTotal'].sum())
    new = new.append(datedf.iloc[:,63:80].sum(), ignore_index = True)
    
    
    date = date + relativedelta(days= 1)

new['date'] = datelst
new['timeCountTotal'] = timelst
new['avgUsedTime'] = avgtime
new['userCountTotal'] = user
```

```python
#각 게임의 366일 평균

average_gameshare = []
average_gameTypeShare = []
gameType = []
p_sum = []
n_sum = []
neu_sum = []
for i in range(len(gamelst)):
    average_gameshare.append(round(df[df.gameName == gamelst[i]].gameShares.sum()/366,2))
    average_gameTypeShare.append(round(df[df.gameName == gamelst[i]].gameTypeShares.sum()/366,2))
    a = df[df.gameName == gamelst[i]]
    gameType.append(a.iloc[1].gameTypeName)
    p_sum.append(df[df.gameName == gamelst[i]].sum()[14:26].sum())
    n_sum.append(df[df.gameName == gamelst[i]].sum()[26:38].sum())
    neu_sum.append(df[df.gameName == gamelst[i]].sum()[38:50].sum())
    
avdf = pd.DataFrame({'gameName': gamelst, 
                    'gameshare': average_gameshare,
                     
                    'gameTypeshare': average_gameTypeShare,
                    'gameType': gameType,
                    'Positive': p_sum,
                    'Negative':n_sum,
                    'Neutral': neu_sum})
avdf
```
