```python
#게임 이름 excel 파일형식으로
lst = list(pf['gameName'])
for i in range(len(lst)):
    lst[i] = ''.join(char for char in lst[i] if char.isalnum())
pf['gameName'] = lst
```

```python
#날짜 바꾸기 20201231 형식으로
lst = list(pf['date'])
for i in range(len(lst)):
    lst[i] = ''.join(char for char in lst[i] if char.isalnum())
    lst[i] = int(lst[i])
pf['date'] = lst 
```

```python
path = "C:/Users/Hannah Kim/Desktop/프로시딩준비/LucyRaw2/*.xlsx"
```

긍정 dataframe = p

```python
p = pd.DataFrame()
for fname in glob.glob(path):

    name = fname.split('_g_')[1]
    name = ''.join(char for char in name if char.isalnum())
    name = name.replace("xlsx", "")
    
    df = pd.read_excel(fname, sheet_name = '긍정')
    df['gameName'] = name
    df = df.rename(columns = {'구분':'date'})
    
    
    p = p.append(df)
        
```

```python
p
```

```python
avdf = pd.read_csv("avdf.csv")
avdf
```

```python
onesixty = list(avdf['gameName'])
plst = list(p['gameName'])
zerolst = []
for x in onesixty:
    if x not in plst:
        zerolst.append(x)
zerolst
```

```python
l = ['블로그', '카페', '커뮤니티', '언론', '페이스북', '정부_공공', '인스타그램', '카카오스토리',
       '기업_단체', '지식인', '트위터', '유투브']
for i in range(len(zerolst)):
    zero = pd.DataFrame()
    zero['date'] = p[p.gameName == '워록']['date']
    for col in l:
        zero[col] = 0
    zero['gameName'] = zerolst[i]
    p = p.append(zero)

```

부정 dataframe = n

```python
n = pd.DataFrame()
for fname in glob.glob(path):

    name = fname.split('_g_')[1]
    name = ''.join(char for char in name if char.isalnum())
    name = name.replace("xlsx", "")
    
    df = pd.read_excel(fname, sheet_name = '부정')
    df['gameName'] = name
    df = df.rename(columns = {'구분':'date'})
    
    
    n = n.append(df)
```

```python
l = ['블로그', '카페', '커뮤니티', '언론', '페이스북', '정부_공공', '인스타그램', '카카오스토리',
       '기업_단체', '지식인', '트위터', '유투브']
for i in range(len(zerolst)):
    zero = pd.DataFrame()
    zero['date'] = p[p.gameName == '워록']['date']
    for col in l:
        zero[col] = 0
    zero['gameName'] = zerolst[i]
    n = n.append(zero)
```

중립 데이터 프레임 = neu

```python
neu = pd.DataFrame()
for fname in glob.glob(path):

    name = fname.split('_g_')[1]
    name = ''.join(char for char in name if char.isalnum())
    name = name.replace("xlsx", "")
    
    df = pd.read_excel(fname, sheet_name = '중립')
    df['gameName'] = name
    df = df.rename(columns = {'구분':'date'})
    
    
    neu = neu.append(df)
```

```python
l = ['블로그', '카페', '커뮤니티', '언론', '페이스북', '정부_공공', '인스타그램', '카카오스토리',
       '기업_단체', '지식인', '트위터', '유투브']
for i in range(len(zerolst)):
    zero = pd.DataFrame()
    zero['date'] = p[p.gameName == '워록']['date']
    for col in l:
        zero[col] = 0
    zero['gameName'] = zerolst[i]
    neu = neu.append(zero)
```

```python
pf = pd.merge(pf, p, on=['date', 'gameName'])
```

```python
pf = pd.merge(pf, n, on=['date', 'gameName'])
```

```python
pf = pd.merge(pf, neu, on=['date', 'gameName'])
```

```python
pf = pf.drop("Unnamed: 0", axis = 1)
```

```python
pf.to_csv("merged_lucy_0818.csv")
```
