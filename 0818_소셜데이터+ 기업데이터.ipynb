# 각 게임마다

```python

average_gameshare = []
average_gameTypeShare = []
gameType = []
p_sum = []
n_sum = []
neu_sum = []
for i in range(len(gamelst)):
    average_gameshare.append(round(df[df.gameName == gamelst[i]].gameShares.sum()/366,2))
    average_gameTypeShare.append(round(df[df.gameName == gamelst[i]].gameTypeShares.sum()/366,2))
    a = df[df.gameName == gamelst[i]]
    gameType.append(a.iloc[1].gameTypeName)
    p_sum.append(df[df.gameName == gamelst[i]].sum()[14:26].sum())
    n_sum.append(df[df.gameName == gamelst[i]].sum()[26:38].sum())
    neu_sum.append(df[df.gameName == gamelst[i]].sum()[38:50].sum())
avdf = pd.DataFrame({'gameName': gamelst, 
                    'gameshare': average_gameshare,
                     
                    'gameTypeshare': average_gameTypeShare,
                    'gameType': gameType,
                    'Positive': p_sum,
                    'Negative':n_sum,
                    'Neutral': neu_sum})
avdf
```
