```python
#gov 포함
average_gameshare = []
average_gameTypeShare = []
gameType = []
p_sum = []
n_sum = []
neu_sum = []
for i in range(len(gamelst)):
    average_gameshare.append(round(df[df.gameName == gamelst[i]].gameShares.sum()/366,2))
    average_gameTypeShare.append(round(df[df.gameName == gamelst[i]].gameTypeShares.sum()/366,2))
    a = df[df.gameName == gamelst[i]]
    gameType.append(a.iloc[1].gameTypeName)
    p_sum.append(df[df.gameName == gamelst[i]].sum()[14:25].sum())
    n_sum.append(df[df.gameName == gamelst[i]].sum()[25:36].sum())
    neu_sum.append(df[df.gameName == gamelst[i]].sum()[36:47].sum())
avdf = pd.DataFrame({'gameName': gamelst, 
                    'gameshare': average_gameshare,
                     
                    'gameTypeshare': average_gameTypeShare,
                    'gameType': gameType,
                    'Positive': p_sum,
                    'Negative':n_sum,
                    'Neutral': neu_sum})
avdf
```

```python
plt.figure(figsize=(16, 6))

heatmap = sns.heatmap(avdf.corr(), vmin=-1, vmax=1, annot=True)
# Give a title to the heatmap. Pad defines the distance of the title from the top of the heatmap.
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12)
```

```python
from scipy.stats import pearsonr
import pandas as pd

def calculate_pvalues(df):
    df = df.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    for r in df.columns:
        for c in df.columns:
            pvalues[r][c] = round(pearsonr(df[r], df[c])[1], 20)
    return pvalues


calculate_pvalues(avdf) 
```
